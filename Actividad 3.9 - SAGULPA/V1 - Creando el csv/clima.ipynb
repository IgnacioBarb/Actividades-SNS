```python
import pandas as pd
```

```python
datos = pd.read_csv('../rawData/climadiciembre2019.csv')
```

```python
datos
```

```python
#nos quedamos con los campos que nos interesan
df = datos[['Hora', 'Tem', 'Hum', 'fecha']].dropna()
df.reset_index(drop=True, inplace=True)
df
```

```python
#Damos el formato deseado a los datos
dias = []
for i in range(len(df)):
    df['Hora'].iloc[i] = int(df['Hora'].iloc[i].split(':')[0])
    df['Tem'].iloc[i] = int(df['Tem'].iloc[i].replace('°',''))
    df['Hum'].iloc[i] = int(df['Hum'].iloc[i].replace('%',''))
    dia = df.iloc[i,3].split(' ')[1]
    dias.append(int(dia)) 
df['Dia'] = dias
df.drop(columns='fecha', inplace=True)
```

```python
df
```

```python
#Nos quedamos con un solo registro por dia y hora 
eliminar = []
regs = {}
for dia in range(1,32):
    regs[dia] = []

for i in range(len(df)):
    dia = df['Dia'].iloc[i]
    hora = df['Hora'].iloc[i]

    if hora in regs[dia]:
        eliminar.append(i)
    else:
        regs[dia].append(hora)
    
    

df.drop(eliminar, axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.to_csv('../csv/CLIMA-DICIEMBRE-2019.csv', index=False)
```
